Feature Chosen: User Session Duration

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, boxcox
from sklearn.preprocessing import PowerTransformer

# Step 1: Simulate User Session Duration data (skewed)
np.random.seed(42)
session_duration = np.random.exponential(scale=10, size=1000)  # Skewed data

# Step 2: Check normality of Original data
stat, p = shapiro(session_duration)
print(f"Original Data - Shapiro-Wilk p-value: {p:.4f}")
sns.histplot(session_duration, kde=True)
plt.title("Original Session Duration")
plt.xlabel("Minutes")
plt.ylabel("Frequency")
plt.show()
```

```python
# Log Transformation

log_data = np.log1p(session_duration)  # log1p to avoid log(0)

stat, p = shapiro(log_data)
print(f"Log Transformed - Shapiro-Wilk p-value: {p:.4f}")
sns.histplot(log_data, kde=True)
plt.title("Log Transformed")
plt.show()
```

```python
# Sqrt Transformed

sqrt_data = np.sqrt(session_duration)

stat, p = shapiro(sqrt_data)
print(f"Square Root Transformed - Shapiro-Wilk p-value: {p:.4f}")
sns.histplot(sqrt_data, kde=True)
plt.title("Square Root Transformed")
plt.show()
```

```python
# Box-cox Transformed
# Box-Cox only works with strictly positive data
boxcox_data, lambda_val = boxcox(session_duration)

stat, p = shapiro(boxcox_data)
print(f"Box-Cox Transformed - Shapiro-Wilk p-value: {p:.4f}, Lambda: {lambda_val:.4f}")
sns.histplot(boxcox_data, kde=True)
plt.title("Box-Cox Transformed")
plt.show()
```
