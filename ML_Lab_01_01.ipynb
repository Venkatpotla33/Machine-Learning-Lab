Task 1
Feature 2 : Sepal Height
we Apply log,Square Root

```python
# Load the Iris dataset
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
print(df)
```

```python
# Apply Transfomation to a Sepeal Height

import numpy as np
from scipy.stats import boxcox
log_transformed = np.log1p(df['sepal length (cm)'])
sqrt_transformed = np.sqrt(df['sepal length (cm)'])
boxcox_transformed, _ = boxcox(df['sepal length (cm)'] + 1e-6)


# Check normality using Shapiro-Wilk test
from scipy.stats import shapiro
print(shapiro(df['sepal length (cm)'].sample(100)))
print(shapiro(pd.Series(log_transformed).sample(100)))
print(shapiro(pd.Series(sqrt_transformed).sample(100)))
print(shapiro(pd.Series(boxcox_transformed).sample(100)))

```

```python
# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(log_transformed, kde=True)
plt.title('Log-transformed Sepal length')
plt.show()

print(shapiro(df['sepal length (cm)'].sample(100)))
print(shapiro(pd.Series(log_transformed).sample(100)))

import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(sqrt_transformed, kde=True)
plt.title('sqrt_transformed Sepal length')
plt.show()
print(shapiro(df['sepal length (cm)'].sample(100)))
print(shapiro(pd.Series(sqrt_transformed).sample(100)))

import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(boxcox_transformed, kde=True)
plt.title('boxcox_transformed sepel length')
plt.show()
print(shapiro(df['sepal length (cm)'].sample(100)))
print(shapiro(pd.Series(boxcox_transformed).sample(100)))
```
