# Task 02

Find the best feature combinations to yield highest classification performance.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your cleaned file from Google Drive link
file_path = 'https://drive.google.com/uc?id=1Yue9kAiRQsbSHzdpSoBfhX9-Xc_c5hKr'
df = pd.read_csv(file_path)

print(df.shape)
print(df.head())

# Select only numerical columns for plotting where needed
numerical_df = df.select_dtypes(include=['number'])

# Pair Plot
print("\nGenerating Pair Plot...")
if numerical_df.shape[1] > 1:
    sns.pairplot(numerical_df)
    plt.suptitle("Pair Plot of Numerical Features", y=1.02)
    plt.show()
else:
    print("Not enough numerical columns for pair plot.")


# Box Plot
print("\nGenerating Box Plot...")
if numerical_df.shape[1] > 0:
    plt.figure(figsize=(14,6))
    sns.boxplot(data=numerical_df)
    plt.title("Box Plot of Numerical Features")
    plt.xticks(rotation=45)
    plt.show()
else:
    print("No numerical columns for box plot.")


# Violin Plot
print("\nGenerating Violin Plot...")
if numerical_df.shape[1] > 0:
    plt.figure(figsize=(14,6))
    sns.violinplot(data=numerical_df)
    plt.title("Violin Plot of Numerical Features")
    plt.xticks(rotation=45)
    plt.show()
else:
    print("No numerical columns for violin plot.")


# Correlation Heatmap
print("\nGenerating Correlation Heatmap...")
if numerical_df.shape[1] > 1:
    plt.figure(figsize=(10,8))
    sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title("Correlation Heatmap of Numerical Features")
    plt.show()
else:
    print("Not enough numerical columns for correlation heatmap.")
```
